```python
import nest_asyncio
nest_asyncio.apply()

import os

import rasa
from rasa.jupyter import chat
import rasa.data as data

config = "config.yml"
training_files = "data/"
domain = "domain.yml"
output = "models/"
model_path = rasa.train(domain, config, [training_files], output)

def chat_with_rasa():
    return chat(model_path)

def test_models():
    
    stories_dir, nlu_data_dir = data.get_core_nlu_directories(training_files)
    return rasa.test(model_path, stories_dir, nlu_data_dir)


# writting results of the core evaluation to a file called results
def test_reports():
    if os.path.isfile("errors.json"):
        print("NLU Errors")
        print(open("errors.json").read())
    else:
        print("No NLU errors.")
    
# writting NLU errors to errors.json

    if os.path.isdir("results"):
        print("\n")
        print("Core Errors:")
        print(open("results/failed_stories.md").read())
```

```python
chat_with_rasa()
test_models()
test_reports()
```
